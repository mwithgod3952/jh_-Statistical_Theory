```python
import math
import matplotlib.pyplot as plt

def binomial_distribution(n, i, p):
    a = math.factorial(n)/(math.factorial(i)*math.factorial(n - i))
    b = math.pow(p, i) * math.pow(1-p, n -i)
    return a*b

def graph(n):
    results = []
    for i in range(1, 20):
        calculation = binomial_distribution(n, i, 1/6)
        results.append(calculation)
    return results

# Visualization
result_nth = []
for n in range(20, 60, 10):
    plt.plot(range(0, len(graph(n)), 1), graph(n))   
```
